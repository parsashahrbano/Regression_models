**Linear Regression**

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin = lin_reg.fit(X_train,y_train)


test_pred = lin.predict(X_test)
train_pred = lin.predict(X_train)

scoretrain = lin.score(X_train,y_train)
scoretest  = lin.score(X_test,y_test)

print('For linear regression')
from sklearn.metrics import mean_squared_error
rmse_test = mean_squared_error(y_test , test_pred , squared=False)
rmsep_test = np.sqrt(np.mean(np.square(((y_test - test_pred) / y_test)), axis=0))
rmsep_train = np.sqrt(np.mean(np.square(((y_train - train_pred) / y_train)), axis=0))
rmse_train = mean_squared_error(y_train , train_pred, squared=False)
print('Test set evaluation:\n_____________________________________')
#print_evaluate(y_test, test_pred)
print('The Root Mean Square Error(RMSE) is: ',rmsep_test)
print('The R^2 is: ',scoretest)

print('Train set evaluation:\n_____________________________________')
#print_evaluate(y_train, train_pred)
print('The Root Mean Square Error(RMSE) is: ',rmsep_train)
print('The R^2 is: ',scoretrain)
```
